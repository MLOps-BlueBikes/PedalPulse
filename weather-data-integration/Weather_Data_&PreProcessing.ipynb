```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2

# Haversine function to calculate the distance between two lat-long pairs
def haversine(lat1, lon1, lat2, lon2):
    # Convert lat/long from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Radius of Earth in kilometers (mean radius)
    R = 6371.0
    return R * c
```

```python

# Create a new column in station_data for the closest weather station
station_data['Closest Weather Station'] = ''
station_data['Weather Station ID'] = ''

# Loop through each station
for i, station in station_data.iterrows():
    station_lat = station['Lat']
    station_lon = station['Long']

    # Compute distances to all weather stations
    distances = []
    for j, weather_station in weather_station_data.iterrows():
        weather_lat = weather_station['Approx. Latitude']
        weather_lon = weather_station['Approx. Longitude']
        distance = haversine(station_lat, station_lon, weather_lat, weather_lon)
        distances.append(distance)

    # Find the index of the minimum distance
    min_distance_index = np.argmin(distances)

    # Assign the closest weather station
    closest_weather_station = weather_station_data.iloc[min_distance_index]
    station_data.at[i, 'Closest Weather Station'] = closest_weather_station['Weather Station Name']
    station_data.at[i, 'Weather Station ID'] = closest_weather_station['Identifier']
```

```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By


# Set up the WebDriver with Chrome options
def set_up_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Ensures Chrome runs without a UI
    options.add_argument('--disable-gpu')  # Disable GPU (helps in some headless cases)
    options.add_argument('--no-sandbox')  # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems
    options.add_argument('--remote-debugging-port=9222')  # Open port for debugging

    driver = webdriver.Chrome(options=options)
    return driver


# Function to scrape weather data for a given date
def scrape_wunderground_data(date):
    driver = set_up_driver()

    # URL format based on the provided link
    url = f"https://www.wunderground.com/history/daily/us/ma/east-boston/KBOS/date/{date}"
    driver.get(url)

    # Wait for the page to load completely
    time.sleep(5)

    # Extract the page source
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # 1st Table: Regular weather data (time series)
    table = soup.find('table', {'class': 'ng-star-inserted'})
    if not table:
        print(f"No weather data table found for {date}")
        return None

    rows = table.find_all('tr')
    weather_data = []

    for row in rows[1:]:  # Skip the header row
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        weather_data.append(cols)

    # Define columns for the regular weather data
    columns = ['Time', 'Temperature (Â°F)', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.', 'Condition']

    df = pd.DataFrame(weather_data, columns=columns[:len(weather_data[0])])  # Adjust based on actual columns in the table

    # 2nd Table: High, Low, and Average Temperatures with "Actual", "Historic Avg", "Record"
    temp_table = soup.find_all('table')[1]  # Assuming this is the second table on the page; adjust if needed
    temp_rows = temp_table.find_all('tr')

    temp_data = []
    for row in temp_rows[1:]:  # Skip the header row
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        temp_data.append(cols)

    # Extract the high, low, and average temp columns
    temp_df = pd.DataFrame(temp_data, columns=['Actual', 'Historic Avg', 'Record'])

    # Transpose the temp_df to merge with time-series data (if necessary)
    temp_df = temp_df.set_index('Metric').T  # Make 'Metric' the index, and transpose to have one row with 'Actual', 'Historic Avg', etc.

    driver.quit()

    # Combine both DataFrames into one final output
    final_df = pd.concat([df, temp_df.reset_index(drop=True)], axis=1)

    return final_df


# Function to loop through multiple dates and save data
def scrape_multiple_days(start_date, end_date):
    # Create a date range
    date_range = pd.date_range(start=start_date, end=end_date)

    all_data = []

    for date in date_range:
        date_str = date.strftime('%Y-%m-%d')
        print(f"Scraping data for {date_str}")
        df = scrape_wunderground_data(date_str)

        if df is not None:
            df['Date'] = date_str  # Add the date to each row
            all_data.append(df)

        # Avoid being blocked
        time.sleep(2)

    # Combine all data into a single DataFrame
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        # Save to CSV
        final_df.to_csv('weather_data.csv', index=False)
        print("Data saved to 'weather_data.csv'")
    else:
        print("No data was scraped.")


# Scrape data from Jan 1, 2023, to Jan 31, 2023 as an example
scrape_multiple_days('2023-01-01', '2023-01-31')
```

```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up the WebDriver with Chrome options
def set_up_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Run in headless mode
    options.add_argument('--disable-gpu')  # Disable GPU rendering (helps in headless)
    options.add_argument('--no-sandbox')  # Bypass OS security model, required in some headless environments
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems
    options.add_argument('--remote-debugging-port=9222')  # Allow remote debugging

    driver = webdriver.Chrome(options=options)
    return driver

# Function to scrape weather data for a given date using XPath
def scrape_wunderground_data(date):
    driver = set_up_driver()

    url = f"https://www.wunderground.com/history/daily/us/ma/east-boston/KBOS/date/{date}"
    driver.get(url)

    try:
        # Wait for the table to be present using the XPath
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/div/div[2]/table'))
        )
        time.sleep(2)  # Give extra time for all elements to load

        # Find the table using the XPath
        table = driver.find_element(By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/div/div[2]/table')

        # Parse the page source
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Extract the table from the page source using the same XPath
        table = soup.find('table')  # Modify if more specific selection is needed

        # Check if the table exists
        if not table:
            print(f"No table found for date: {date}")
            return None

        # Find all the rows in the table
        rows = table.find_all('tr')

        # Extract data from each row
        weather_data = []
        for row in rows[1:]:  # Skip the header row
            cols = row.find_all('td')
            cols = [col.text.strip() for col in cols]
            weather_data.append(cols)

        # Define column names (make sure these match the actual table headers)
        columns = ['Time', 'Temperature', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.', 'Condition']

        # Create a DataFrame from the data
        df = pd.DataFrame(weather_data, columns=columns)

        return df

    except Exception as e:
        print(f"Error scraping data for {date}: {e}")
        return None
    finally:
        driver.quit()

# Function to scrape data across a range of dates
def scrape_multiple_days(start_date, end_date):
    date_range = pd.date_range(start=start_date, end=end_date)
    all_data = []

    for date in date_range:
        date_str = date.strftime('%Y-%m-%d')
        print(f"Scraping data for {date_str}")
        df = scrape_wunderground_data(date_str)
        if df is not None:
            df['Date'] = date_str  # Add the date column
            all_data.append(df)

        # Be mindful of rate-limiting and avoid being blocked by pausing between requests
        time.sleep(2)

    if all_data:


        # Concatenate all DataFrames into a single one
        final_df = pd.concat(all_data, ignore_index=True)
        final_df.to_csv('weather_data.csv', index=False)
        print("Data saved to 'weather_data.csv'")
    else:
        print("No data was scraped.")

# Example: Scrape weather data for January 2023
scrape_multiple_days('2023-01-01', '2023-01-03')
```
